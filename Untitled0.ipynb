```python
import os
import pandas as pd
import numpy as np
import kagglehub
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report

# 1. DOWNLOAD & LOAD DATA
print("🚀 Downloading dataset...")
path = kagglehub.dataset_download("aryanhirlekar/ckdchronic-kidney-disease-dataset-with-stages")
csv_files = [f for f in os.listdir(path) if f.endswith('.csv')]
df = pd.read_csv(os.path.join(path, csv_files[0]))

# 2. DATA CLEANING
print("🧼 Cleaning data...")
# Drop columns that are irrelevant or leakage
cols_to_drop = ['id', 'ckd_pred']
df = df.drop(columns=[c for c in cols_to_drop if c in df.columns])

# Handle missing values
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna(df[col].mode()[0])
    else:
        df[col] = df[col].fillna(df[col].median())

# 3. ENCODING & SCALING
print("🔢 Encoding features...")
le_features = LabelEncoder()

# Confirm the target column name
target_col = 'ckd_stage'  # Update if different in your dataset
y_raw = df[target_col]
X_raw = df.drop(columns=[target_col])

# Encode categorical features
for col in X_raw.select_dtypes(include=['object']).columns:
    X_raw[col] = le_features.fit_transform(X_raw[col].astype(str))

# Encode target variable
le_target = LabelEncoder()
y = le_target.fit_transform(y_raw)

# Split dataset
X_train_raw, X_test_raw, y_train, y_test = train_test_split(
    X_raw, y, test_size=0.2, random_state=42)

# Scale features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train_raw)
X_test = scaler.transform(X_test_raw)

# 4. TRAIN DEEP LEARNING MODEL (MLP)
print("🧠 Training Deep Learning Model (MLP)...")
mlp = MLPClassifier(
    hidden_layer_sizes=(128, 64, 32),
    activation='relu',
    solver='adam',
    alpha=0.0001,
    batch_size='auto',
    learning_rate='adaptive',
    max_iter=1000,
    random_state=42,
    verbose=False
)

mlp.fit(X_train, y_train)

# 5. EVALUATION
print("\n✅ Model Evaluation:")
y_pred = mlp.predict(X_test)
print(classification_report(y_test, y_pred, target_names=[str(c) for c in le_target.classes_]))

# 6. SAVE FOR DEPLOYMENT
print("💾 Saving model artifacts...")
joblib.dump(mlp, 'ckd_deep_model.pkl')
joblib.dump(scaler, 'scaler.pkl')
joblib.dump(le_target, 'target_encoder.pkl')

print("\n✨ Done! You have 3 files ready for deployment.")
```
